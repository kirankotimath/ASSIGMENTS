```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
!pip install mlxtend
```

```python
book=pd.read_csv("C:\\Users\\kiran\\Downloads\\book.csv")
book
```

# 1 Association rules with 10% support and 50% confidence 

```python
dataset1=apriori(book,min_support=0.1,use_colnames=True)
dataset1.shape
```

```python
rule1=association_rules(dataset1,metric='confidence',min_threshold=0.5)
```

```python
rule1.shape
```

```python
# AS WE KNOW THAT LIFT RATIO SHOULD BE GREATER THEN 1 
```

```python
rule1=rule1[rule1.lift>1]
rule1
```

```python
# the number of rules is 49
```

```python
rule1[['support','confidence','lift']].hist()
```

```python
plt.scatter(rule1.support,rule1.confidence)
plt.xlabel('support')
plt.ylabel('confidence')
```

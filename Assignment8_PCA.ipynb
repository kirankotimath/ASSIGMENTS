```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("C:\\Users\\kiran\\Downloads\\wine.csv")
wine
```

```python
wine.Type.value_counts()
```

```python
data1=wine.iloc[:,1:]
data = wine.iloc[:,1:].values
```

```python
data_norm = scale(data)
data_norm
```

```python
pcs = PCA()
pcs_values = pcs.fit_transform(data_norm)
pcs_data = pd.DataFrame(pcs_values)
pcs_data
```

```python
pcs.explained_variance_ratio_
```

```python
var = (pcs.explained_variance_ratio_)*100
pd.DataFrame(var)
```

# Performing Heirarchial Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

dist = sch.linkage(pcs_data.iloc[:,:3],method='complete')
#create a dendrogram :-

dendrogram = sch.dendrogram(dist)
```

```python
# create a cluster :-

cluster = AgglomerativeClustering(linkage='ward',affinity='euclidean',n_clusters=3)
y_cluster = cluster.fit_predict(pcs_data.iloc[:,:3])
pd.DataFrame(y_cluster).value_counts()
```

```python
data1['h_cluster']=pd.Series(y_cluster)
data1
```

```python
result_h_cluster = data1.groupby(data1.h_cluster).mean()
result_h_cluster
```

```python
result_original = wine.iloc[:,1:].groupby(wine.Type).mean()
result_original
```

# now we will perform KMeans Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
pcs_data1 = pcs_data.iloc[:,:3]
```

```python
kmeans_cluster = KMeans(n_clusters=3)
kmeans_cluster.fit(pcs_data1)
```

```python
kmeans_cluster.labels_
```

```python
k = list(range(1,6))
k
TWSS = []
for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pcs_data1)
    WSS = []
    for j in range(i):
        WSS.append(sum(cdist(pcs_data1.iloc[kmeans_cluster.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,pcs_data1.shape[1]),"euclidean")))
    print("WSS when k is ", i , sum(WSS))
    TWSS.append(sum(WSS))
```

```python
plt.plot(k,TWSS,"ro-")
```

```python
data1['Cluster_ID'] = pd.Series(kmeans_cluster.labels_)
data1
```

```python
data1.Cluster_ID.value_counts()
```

```python
data2 = data1.iloc[:,:-1]
data2.head()
```

```python
result_kmeans_cluster = data2.groupby(data1.Cluster_ID).mean()
```

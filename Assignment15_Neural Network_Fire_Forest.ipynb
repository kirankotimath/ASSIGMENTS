```python
import pandas as pd
import numpy as np
from tensorflow import keras 
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, load_model
from sklearn.preprocessing import LabelEncoder
```

```python
seed = 7
np.random.seed(seed)
data = pd.read_csv("C:\\Users\\kiran\\Downloads\\forestfires (1).csv")
data
```

```python
label_encoder = LabelEncoder()
```

```python
data['fire_size'] = label_encoder.fit_transform(data.size_category)
```

```python
# split the data :-

X = data.iloc[:,2:-2].values
Y = data.iloc[:,-1].values
```

```python
# we will standardize our data 
from sklearn.preprocessing import StandardScaler

a = StandardScaler()
a.fit(X)
X_std = a.transform(X)
```

```python
# now create a neural-network model :-

model = Sequential()
model.add(layers.Dense( 28, input_dim = 28, activation='relu'))
model.add(layers.Dense( 15, activation='relu'))    # this is the hidden layer
model.add(layers.Dense( 1, activation='sigmoid'))
```

```python
# now we will complile our model :-

model.compile (loss='hinge', optimizer='adam', metrics=['accuracy'])
```

```python
# now we will fit our data into model :-

model.fit(X,Y, validation_split = 0.3, epochs=300, batch_size=15)
```

```python
score = model.evaluate(X,Y)
```

```python
print(score[1]*100)
print(model.metrics_names[1])
```

# Tuning Hyperparameter :- Batch Size and Epoch

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def tuning_model():
    model = Sequential()
    model.add(layers.Dense( 28, input_dim = 28, activation='relu'))
    model.add(layers.Dense( 15, activation='relu'))    # this is the hidden layer
    model.add(layers.Dense( 1, activation='sigmoid'))
    
    model.compile (loss='hinge', optimizer='adam', metrics=['accuracy'])
    
    return model
```

```python
# create a model:-
model1 = KerasClassifier(build_fn=tuning_model)

# give the parameeters:-
batch_size = [10,30,50]
epochs = [200,400,500]

grid_param = dict(batch_size=batch_size , epochs=epochs )

# build the grid and fit the data :-
grid = GridSearchCV(estimator=model1, param_grid=grid_param)
result = grid.fit(X,Y)
```
